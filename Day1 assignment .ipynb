```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#line chart
x = [1,2,3,4,5]
y = [0,2,4,6,8]

plt.plot(x,y, label='xy', color='green', marker='v', linestyle='dashed', markersize=15, markeredgecolor='g', linewidth=5)

plt.title('Assignment line chart 1')
plt.legend()

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xticks([1,2,3,4,5])
plt.yticks([0,2,4,6,8])

plt.savefig('mygraph.png', dpi=300)

plt.show()
```
